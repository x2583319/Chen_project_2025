```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# SVM (using cross-validation)
svm_clf = SVC(kernel='linear')
svm_scores = cross_val_score(svm_clf, X_bin_pca, y_binary, cv=3)
print("SVM accuracy:", svm_scores.mean())

# KNN (manual split to avoid cross_val_score bug)
X_train, X_test, y_train, y_test = train_test_split(X_bin_pca, y_binary, test_size=0.33, random_state=42)

knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, y_train)
y_pred = knn_clf.predict(X_test)
knn_acc = accuracy_score(y_test, y_pred)
print("KNN accuracy (manual split):", knn_acc)
```
